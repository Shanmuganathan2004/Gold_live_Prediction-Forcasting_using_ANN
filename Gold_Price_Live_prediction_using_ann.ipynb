```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Step 1: Data Collection
# Fetch historical gold prices (symbol 'GC=F' for gold futures in USD)
start_date = '2020-01-01'
end_date = datetime.today().strftime('%Y-%m-%d')
gold_data = yf.download('GC=F', start=start_date, end=end_date)

# Step 2: Data Preprocessing
# Select the 'Close' price and drop missing values
gold_prices = gold_data[['Close']].dropna()

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(gold_prices)

# Prepare the dataset for the LSTM model
def create_dataset(dataset, time_step=1):
    X, Y = [], []
    for i in range(len(dataset) - time_step - 1):
        a = dataset[i:(i + time_step), 0]
        X.append(a)
        Y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(Y)

time_step = 100
X, y = create_dataset(scaled_data, time_step)

# Reshape input to be [samples, time steps, features] which is required for LSTM
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split the data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Step 3: Model Building
# Build the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, batch_size=64, epochs=50, validation_data=(X_test, y_test))

# Step 4: Prediction
# Predicting on the test data
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Transform back to original form
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)

# Calculate RMSE
import math
from sklearn.metrics import mean_squared_error
train_rmse = math.sqrt(mean_squared_error(y_train, train_predict))
test_rmse = math.sqrt(mean_squared_error(y_test, test_predict))
print(f'Train RMSE: {train_rmse}')
print(f'Test RMSE: {test_rmse}')

# Plotting
# Shift train predictions for plotting
train_plot = np.empty_like(scaled_data)
train_plot[:, :] = np.nan
train_plot[time_step:len(train_predict) + time_step, :] = train_predict

# Shift test predictions for plotting
# Shift test predictions for plotting
test_plot = np.empty_like(scaled_data)
test_plot[:, :] = np.nan
# The issue was with the slicing of test_plot.
# The original calculation was incorrect and resulted in a slice with a smaller shape than test_predict.
# The corrected calculation ensures the slice has the same number of rows as test_predict.
test_plot[len(train_predict) + time_step:len(train_predict) + time_step + len(test_predict), :] = test_predict
# Plot baseline and predictions
plt.figure(figsize=(10, 4))
plt.plot(scaler.inverse_transform(scaled_data), label='Actual Gold Prices')
plt.plot(train_plot, label='Train Prediction')
plt.plot(test_plot, label='Test Prediction')
plt.xlabel('Days')
plt.ylabel('Gold Price (USD)')
plt.title('Gold Price Prediction using LSTM')
plt.legend()
plt.show()

# Predicting future prices
# Get the last 'time_step' days from the dataset
last_days = scaled_data[-time_step:]
# Create an empty list to store the predictions
future_predictions = []
# Predicting the next 30 days
for _ in range(30):
    # Reshape the data to match the LSTM input shape
    last_days_reshaped = last_days.reshape(1, time_step, 1)
    # Predict the next day
    next_day_pred = model.predict(last_days_reshaped)
    # Append the prediction to the list
    future_predictions.append(next_day_pred[0, 0])
    # Update the last_days array by appending the prediction and removing the first day
    last_days = np.append(last_days, next_day_pred)[1:]

# Transform the predictions back to the original scale
future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

# Create a date range for the future predictions
last_date = gold_prices.index[-1]
future_dates = [last_date + timedelta(days=i) for i in range(1, 31)]

# Plotting future predictions
plt.figure(figsize=(10, 4))
plt.plot(gold_prices.index, gold_prices['Close'], label='Historical Gold Prices')
plt.plot(future_dates, future_predictions, label='Future Predictions', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Gold Price (USD)')
plt.title('Gold Price Prediction for the Next 30 Days')
plt.legend()
plt.show()
```
